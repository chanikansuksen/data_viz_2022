## ต่อในแนวแกน  X

```
personal_data = data_df[['No.', 'sex', 'age', 'nationality']]
personal_data.head()
```

```
covid_data = data_df [['No.', 'announce_date', 'province_of_isolation', 'province_of_onset', 'risk']]
covid_data.head()
```

```
covid_data = data_df [['No.', 'announce_date', 'province_of_isolation', 'province_of_onset', 'risk']].sample(frac=1)  # สลับ row สุ่มมั่ว โดยใช้ .sample(frac=1)
covid_data.head()
```
