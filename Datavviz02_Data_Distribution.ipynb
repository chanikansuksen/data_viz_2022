2. วาด boxplot เปรียบเทียบการกระจายของข้อมูล PetalLength ของดอก iris ทั้ง 3 ชนิด

เขียน function ที่รับ input เป็น output ของ boxplot แล้วแสดงค่า min, max, q1, q2, q3

```python
ybp = plt.boxplot(df["SepalLength"], vert = False, meanline = True, showmeans=True)
```

```python
def box_vals(ybp) :
        mmin = min(ybp['whiskers'][0].get_xdata())
        mmax = max(ybp['whiskers'][1].get_xdata())
        q1 = max(ybp ['whiskers'][0].get_xdata()) 
        q2 = ybp['medians'][0].get_xdata()[0]
        q3 = min(ybp ['whiskers'][1].get_xdata()) 
        print(f' min เท่ากับ {mmin} \n max เท่ากับ {mmax} \n Q1 เท่ากับ {q1} \n Q2 เท่ากับ {q2} \n Q3 เท่ากับ {q3} ')
```

```python
box_vals(ybp)
```

```python
def detail_boxretern(boxreturn) :
    q1 = boxreturn['boxes'][0].get_ydata()[0]
    q2 = boxreturn['medians'][0].get_ydata()[0]
    q3 = boxreturn['boxes'][0].get_ydata()[2]
    minv = min(boxreturn['whiskers'][0].get_ydata())
    maxv = max(boxreturn['whiskers'][1].get_ydata())
    print(f'min = {minv}, max = {maxv}, q1 = {q1}, q2 = {q2}, q3 = {q3}')
```

```python
bb2 = plt.boxplot(df['SepalLength'], vert = False)
```

```python
detail_boxretern(bb2)
```

```python
def detail_boxretern_v2(boxreturn, vert = True):
    if vert == True :
        q1 = boxreturn['boxes'][0].get_ydata()[0]
        q2 = boxreturn['medians'][0].get_ydata()[0]
        q3 = boxreturn['boxes'][0].get_ydata()[2]
        minv = min(boxreturn['whiskers'][0].get_ydata())
        maxv = max(boxreturn['whiskers'][1].get_ydata())
    else : #vert == False
        q1 = boxreturn['boxes'][0].get_ydata()[0]
        q2 = boxreturn['medians'][0].get_ydata()[0]
        q3 = boxreturn['boxes'][0].get_ydata()[2]
        minv = min(boxreturn['whiskers'][0].get_ydata())
        maxv = max(boxreturn['whiskers'][1].get_ydata())    

    print(f'min = {minv}, max = {maxv}, q1 = {q1}, q2 = {q2}, q3 = {q3}')
```

```python
detail_boxretern_v2(bb2, vert = False)
```
